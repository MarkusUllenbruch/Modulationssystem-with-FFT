```python
import numpy as np
import librosa
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
from scipy import signal
```

# Tiefpassfilter im Zeitbereich Implementierung

```python
def lowpass(samples, w_s, w_c, n):
    """ Tiefpassfilterung mit Verwendung von Hamming-Window
    samples -- Gesampelte Daten die gefiltert werden sollen
    w_s     -- sampling frequency w_s = 1/ T_A
    w_c     -- Grenzfrequenz, Frequenzen > w_c werden gefiltert
    n       -- Ordnung des Filters
    """
    w_c_head = 2 * w_c / w_s
    x = np.zeros(n)
    w = np.zeros(n)
    h = np.zeros(n)
    for k in range(n):
        x[k] = w_c_head * np.sinc(w_c_head * (k - 0.5*n))
        w[k] = 0.54 - 0.46 * np.cos(2 * np.pi * k/n)  
    h = x * w
    convolved = schnelle_faltung(samples, h) #Tiefpass mit eigener schneller Faltung, alternativ: scipy: signal.fftconvolve(X,Y)
    return convolved
```

## Test Tiefpass-Filter mit Audiosignal

```python
def schnelle_faltung(f_arg, g_arg):
    """ Berechne schnelle Faltung von 2 Inputsignalen
    f_arg  --  Inputsignal 1
    g_arg  --  Inputsignal 2
    """

    if f_arg.size >= g_arg.size: # Bezeichne längeres Signal als f
        f = f_arg                # Bezeichne kürzeres Signal als g
        g = g_arg
    else:
        f = g_arg
        g = f_arg
        
    G = g.shape[0]               # Länge Impulsantwort       
    F = 8*G                      # Länge Fenster --> grobe Festlegung, Daumenregel aus Vorlesung
    F = 2**(int(np.log2(F)) + 1) # F>G und 2er-Potenz, es wird die nächste 2-er Potenz berechnet zu 8*G
    
    g = np.pad(g, (0, F-G), 'constant', constant_values=(0,0)) # Füge F-G Nullen ans Ende von g ein
    f = np.pad(f, (G-1, 0), 'constant', constant_values=(0,0)) # Füge G-1 Nullen an Anfang von f ein
    len_F = len(f)         # Länge des zero gepaddeten Signals f
    h = [];                # Initialisiere leeren Array h  
    delta = F - G + 1
    s = 0                  # Shift-Index       
    while True: 
        f_block = [];
        if s+F-1 >= len_F:  #Abbruchbedingung der while-Schleife Bevor es zum "break" kommt, muss noch  verbleibende Signal verarbeitet werden
            signal_remaining = f[s:]  
            part_for_FFT = np.concatenate((signal_remaining, np.zeros(F-len(signal_remaining))) )
            cyc = IFFT(rec_FFT(part_for_FFT)*rec_FFT(g))
            h = np.concatenate((h, cyc[G-1: G+len(signal_remaining)-1]))   ##   G!!! statt G-1
            break
        else:
            f_block = f[s:s+F]
            z =  IFFT( rec_FFT(f_block) * rec_FFT(g) )  
            h = np.concatenate((h, z[G-1: F]))
            s = s + delta
        
    h = np.array(h, dtype=np.float64)    
    return h
```

```python
def rec_FFT(f):
    """ Recursive FFT calculation of samples f """
    f = np.asarray(f)
    n = f.shape[0]
    if n == 1:
        return f
    else:
        f_even = f[::2]
        f_odd = f[1::2]
        g = rec_FFT(f_even)
        u = rec_FFT(f_odd)
        c = np.zeros(n, dtype=complex)
        for k in range(n//2):
            c[k] = g[k] + u[k]*np.exp(-2j*np.pi*k/n)
            c[k+n//2] = g[k] - u[k]*np.exp(-2j*np.pi*k/n)
    return c
```

```python
def IFFT(z):
    """Inverse Recursive FFT calculation of Fourier coefficients z"""
    n = len(z)
    f = IFFT_rec(z)
    return np.real(f)/n
```

```python
def IFFT_rec(z):
    """  """
    z = np.asarray(z, dtype=complex)
    n = z.shape[0]
    z_even = z[::2]
    z_odd = z[1::2]
    
    if n == 1:
        return z
    else:
        g = IFFT_rec(z_even)
        u = IFFT_rec(z_odd)
        c = np.zeros(n, dtype=complex)
        
        for k in range(n//2):
            temp = u[k]*np.exp(2j*np.pi*k/n)
            c[k]      = g[k] + temp
            c[k+n//2] = g[k] - temp
    return c
```

## Test des Tiefpass-Filter mit harmonischen Testsignalen

```python
#Abtastung
def sampling(t_end, sampling_freq):
    sampling = np.linspace(0, t_end, t_end * sampling_freq, endpoint=True)
    return sampling
```

```python
#Sinus Signal
def sine(freq, amp, sampling):
    signal = amp * np.sin(2*np.pi*freq*sampling)
    return signal
```

```python
fs = 1000
print(fs)
Ts = 1/fs
f1 = 2
f2 = 10

fc = 7


time1 = sampling(2, fs)
data1 = sine(f1, 1, time1)
data2 = sine(f2, 1, time1)
data = data1 + data2
```

### Testsignal besteht aus harmonischer Summe unterschiedlicher Frequenzen

```python
plt.plot(time1,data,label='2 Hz')
plt.plot(time1,data1, label='10 Hz')
plt.plot(time1,data2,label='Addition')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.legend(title='')
#plt.savefig('Bilder\Filter_TimeSig.eps') # Für Doku/Präsentation
#plt.savefig('Bilder\Filter_TimeSig.svg') # Für Doku/Präsentation
plt.show()
```

### Gefiltertes Signal

```python
n = 1200
filtered = lowpass(data, fs, fc, n)
time_filter = sampling(len(filtered) * Ts, fs)
plt.plot(time_filter,filtered)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Zeit Verzögerung durch Tiefpass Filter')
#plt.savefig('Bilder\Filter_FilterSig.eps') # Für Doku/Präsentation
#plt.savefig('Bilder\Filter_FilterSig.svg') # Für Doku/Präsentation
plt.show()
```

```python
delta_t = n/2 * Ts
print('Zeitverzögerung durch Lowpass-Filter um t =',delta_t)
```
