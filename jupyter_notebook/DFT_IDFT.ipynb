# DFT und IDFT Implementierung

```python
def DFT(f):
    """Diskrete Fourier Transformation
    f  --  Input Samples
    """
    n = len(f)
    B = np.zeros( (n,n), dtype=complex )
    for k in range(0,n):
        for l in range(0,n):
            B[k,l] = np.exp(2j*np.pi*k*l/n)
    B_ad = np.conjugate(np.transpose(B))
    return np.dot(B_ad, f)
```

```python
def IDFT(z):
    """Inverse Diskrete Fourier Transformation
    z  --  Input Fourier Coefficients
    """
    z = np.asarray(z, dtype=complex)
    n = z.shape[0]
    B = np.zeros( (n,n), dtype=complex )
    for k in range(0,n):
        for l in range(0,n):
            B[k,l] = np.exp(2j*np.pi*k*l/n)
    return np.real(np.dot(B,z))/n
```

# Test DFT & IDFT mit Testsignalen

### DFT mit Testsignalen

```python
#Abtastung
def sampling(t_end, sampling_freq):
    sampling = np.linspace(0, t_end, t_end * sampling_freq, endpoint=False)
    return sampling
```

```python
#Sinus Signal
def sine(freq, amp, sampling):
    signal = amp * np.sin(2*np.pi*freq*sampling)
    return signal
```

```python
f1 = 2
f2 = 4
f3 = 10

fs = 300
t_end = 1

time = sampling(t_end, fs)
data1 = sine(f1, 1, time)
data2 = sine(f2, 1.5, time)
data3 = sine(f3, 2, time)

data = data1 + data2 + data3

plt.plot(time, data)
plt.title('Inputsignal')
plt.show()
```

```python
dft_self = DFT(data)
dft_py = np.fft.fft(data)

diff = dft_py - dft_self
diff_real = np.real(diff)
diff_imag = np.imag(diff)
```

```python
plt.plot(diff_real)
plt.xlabel('Sample')
plt.title('Differenz Realteile')
plt.ylabel('Differenz')
plt.show()
```

```python
plt.plot(diff_imag)
plt.title('Differenz Imaginärteile')
plt.xlabel('Sample')
plt.ylabel('Differenz')
plt.show()
```

### Rücktransformation IDFT

```python
fun = IDFT(dft_self)
diff = fun - data

plt.plot(fun, label='Rekonstruiertes Signal mit IDFT(DFT(X))')
plt.plot(data,'-.', label='Ursprünglches Signal')
plt.ylabel('Signals')
plt.xlabel('Sample')
plt.legend(title='')
plt.show()
```
